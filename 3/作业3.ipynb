```python
df.index=range(0,len(df))
df[:5]
```

```python
df.columns=['名字','类型','城市','地区','地点','评分','评分人数','价格']
df[:5]
```

（3）查看所有类型为“浪漫情侣”的酒店

```python
df[df.类型=='浪漫情侣']
```

（4）查看所有类型为“浪漫情侣”，地区在湾仔的酒店

```python
df[(df.类型=='浪漫情侣')&(df.地区=='湾仔')]
```

（5）查看所有地址在观塘或者油尖旺，评分大于4的酒店

```python
df[((df.地区=='观塘')|(df.地区=='油尖旺'))&(df.评分>4)]
```
