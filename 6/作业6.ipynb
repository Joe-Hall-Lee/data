（1）画出$y=x^{2}+2x+1$在区间[-5,3]的函数图像。

```python
%pylab inline
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
plt.rcParams['font.sans-serif']=['SimHei'] #配置显示中文，否则乱码
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号，如果是mlp画图，则将plt换成mlp
```

```python
x=np.linspace(-5,3,200)
y=x**2+2*x+1
```

```python
plt.figure(figsize=(9,6))
x=np.linspace(-5,3,200)
y=x**2+2*x+1
plt.plot(x,y,'r-')
plt.plot([-1,-1],[0,16],'b--')
plt.xlabel('x',fontsize=18)
plt.ylabel('y',fontsize=18)
plt.title('二次函数',fontsize=30)
plt.show()
```

（2）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.

```python
x=np.linspace(-3.14,3.14,200)
y1=np.sin(x)
y2=np.cos(x)
plt.figure(figsize=(12,6))
plt.subplot(221)
plt.xlabel('x')
plt.plot(x,y1,'b-',x[::20],y1[::20],'k+',linewidth=2.5)
#plt.subplot(1,2,2)
plt.subplot(222)
plt.plot(x,y2,'r-',linewidth=2.5)
plt.xlabel('x')
plt.subplot(223)
plt.plot(x,y2,'k-',linewidth=2.5)
plt.xlabel('x')
plt.subplot(2,2,4)
plt.plot(x,y2,'y-',linewidth=2.5)
plt.xlabel('x')
plt.show()
```

```python
data=df['地区'].value_counts()
x=data.index
y=data.values
plt.figure(figsize=(10,6)) #定义图片大小，长和宽。
plt.bar(x,y,color='r') #color='r'为红色
plt.title('各地区酒店数量',fontsize=20) #定义图片标题，字体大小为20
plt.xlabel('地区',fontsize=18) #定义x轴标题和字体大小
plt.ylabel('酒店数量',fontsize=18) #定义y轴标题和字体大小
#坐标轴大小
plt.tick_params(labelsize=14) #设置坐标轴字体大小
plt.xticks(rotation=90) #将x轴旋转90度显示
#显示网格线
#plt.grid()
#使用text显示数值
for a,b in zip(x,y):
    plt.text(a,b+1,'%.0f' % b,ha='center',va='bottom',fontsize=10)
plt.show()
```

```python
data=df['等级'].value_counts()
label=data.index
y=data.values
y=y/sum(y)
plt.figure(figsize=(10,10))
plt.title('等级占比',fontsize=15)
#plt.pie(y,lebels=label)
#通过函数返回值来定义图像中的字体
patches,l_text,p_text=plt.pie(y,labels=label,autopct = "%3.2f %%", labeldistance=1.1, startangle = 90,pctdistance=0.6,colors = "cmy") #可以自定义颜色
for i in p_text: #设置饼图内部字体
    i.set_size(15)
    i.set_color('k')
for i in l_text: #设置饼图外部字体
    i.set_size(15)
    i.set_color('r')
plt.legend()
plt.show()  
```

（7）画出每个等级酒店评分均值的柱状图。（按照评分均值从小到大排序。）
